# Part 2 - Data ingestion and preprocessing

## Read in data and split into subsets
*   Roughly 68% training, 17% validation and 15% testing data.


```python
(trainX, trainY_original), (testX, testY_original) = keras.datasets.fashion_mnist.load_data()

trainX, valX, trainY_original, valY_original = train_test_split(trainX, trainY_original, test_size=0.2)
```

```python
print("Train X Shape: ", trainX.shape)
print("Val X Shape: ", valX.shape)
print("Test X Shape: ", testX.shape)
```

```python
print("Original Train Y Shape: ", trainY_original.shape)
print("Original Val Y Shape: ", valY_original.shape)
print("Original Test Y Shape: ", testY_original.shape)
```

## Reshape and normalize X

```python
trainX = trainX.reshape(-1, 28, 28, 1)
valX = valX.reshape(-1, 28, 28, 1)
testX = testX.reshape(-1, 28, 28, 1)


trainX = trainX /255.0
valX = valX /255.0
testX = testX /255.0
```

```python
print("Train X Shape: ", trainX.shape)
print("Val X Shape: ", valX.shape)
print("Test X Shape: ", testX.shape)
```

## One hot encoding of class variables

```python
trainY = keras.utils.to_categorical(trainY_original, 10)
valY = keras.utils.to_categorical(valY_original, 10)
testY = keras.utils.to_categorical(testY_original, 10)

print("Encoded Train Y Shape: ", trainY.shape)
print("Encoded Val Y Shape: ", valY.shape)
print("Encoded Test Y Shape: ", testY.shape)
```

## Set Global Variables

```python
class_labels = {0 : 'T-shirt/top',
            1 : 'Trouser',
            2 : 'Pullover',
            3 : 'Dress',
            4 : 'Coat',
            5 : 'Sandal',
            6 : 'Shirt',
            7 : 'Sneaker',
            8 : 'Bag',
            9 : 'Ankle boot'}

imgShape = [28,28,1]
numClasses = 10
classes = list(class_labels.values())
```

## Other functions

```python
def Plot_Training_Progress(history):
    acc = history['accuracy']
    val_acc = history['val_accuracy']
    loss = history['loss']
    val_loss = history['val_loss']
    epochs = range(len(acc))

    fig, axes = plt.subplots(1, 2, figsize=(25,10))
    
    #Plot Accuracy
    axes[0].plot(epochs, acc, 'b-', label='Training Accuracy')
    axes[0].plot(epochs, val_acc, 'g--', label='Validation Accuracy')
    axes[0].set_title('Training and Validation Accuracy')
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Accuracy')
    axes[0].legend(loc='best')

    #Plot Loss
    axes[1].plot(epochs, loss, 'b-', label='Training Loss')
    axes[1].plot(epochs, val_loss, 'g--', label='Validation Loss')
    axes[1].set_title('Training and Validation Loss')
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('Loss')
    axes[1].legend(loc='best')

    return plt
```

# Part 3 - Baseline Model

## Build Model

```python
model = keras.Sequential()
model.add(Conv2D(filters=16,kernel_size=3,activation='relu',input_shape=imgShape))
model.add(Conv2D(filters=16,activation='relu',kernel_size=3))
model.add(MaxPooling2D(pool_size=2))
model.add(Dropout(0.5))
model.add(Conv2D(filters=32,kernel_size=3,activation='relu'))
model.add(Conv2D(32,activation='relu',kernel_size=3))
model.add(MaxPooling2D(pool_size=2))
model.add(Dropout(0.3))
model.add(Flatten())
model.add(Dense(128))
model.add(Dropout(0.2))
model.add(Dense(numClasses, activation='softmax'))
```

## Compile the model

```python
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001),
  loss='categorical_crossentropy',
  metrics=['accuracy']
)
```

```python
keras.utils.plot_model(model, to_file='Baseline_Model_Architecture.png', show_shapes=True, show_layer_names=True)
```

```python
model.summary()
```

## Fit the model

```python
Fashion_MNIST_Baseline_History = model.fit(trainX, trainY, epochs=20, validation_data=(valX, valY))
```

## Plot training and validation accuracy and loss

```python
plt = Plot_Training_Progress(Fashion_MNIST_Baseline_History.history)   
plt.savefig('Baseline_Implementation_Plots.png', dpi=400)    
plt.show()
```

# Part 4 - Hyperparamter Tuning

## Create model and set hyperparameters to tune

```python
class MNIST_Fashion_Hyper(HyperModel):
  def __init__(self, input_shape, num_classes):
    self.input_shape = input_shape
    self.num_classes = num_classes

  def build(self, hp):
    model = keras.Sequential()

    model.add(Conv2D(filters=hp.Choice('Conv_Filters_1',values=[16, 32, 64],default=16),activation='relu',kernel_size=3, input_shape=self.input_shape))
    model.add(Conv2D(filters=hp.Choice('Conv_Filters_2',values=[16, 32, 64],default=16), activation=hp.Choice('Conv_Activation_2',values=['relu', 'sigmoid'],default='relu'),kernel_size=3))
    model.add(MaxPooling2D(pool_size=2))
    model.add(Dropout(rate=hp.Float('Dropout_Rate_1',min_value=0.0,max_value=0.5,default=0.25,step=0.05)))
  
    model.add(Conv2D(filters=hp.Choice('Conv_Filters_3',values=[32, 64, 128],default=32), activation=hp.Choice('Conv_Activation_3',values=['relu', 'sigmoid'],default='relu'),kernel_size=3))
    model.add(Conv2D(filters=hp.Choice('Conv_Filters_4',values=[32, 64, 128],default=64), activation=hp.Choice('Conv_Activation_4',values=['relu', 'sigmoid'],default='relu'),kernel_size=3))
    model.add(AveragePooling2D(pool_size=2))
    model.add(Dropout(rate=hp.Float('Dropout_Rate_2',min_value=0.0,max_value=0.5,default=0.25,step=0.05)))

    model.add(Flatten())
    model.add(Dense(units=hp.Int('Last_Dense_NumUnits',min_value=16,max_value=128,step=16,default=64), activation='relu'))
    model.add(Dropout(rate=hp.Float('Dropout_Rate_3',min_value=0.0,max_value=0.5,default=0.25,step=0.05)))
    model.add(Dense(self.num_classes, activation='softmax'))

    model.compile(optimizer=keras.optimizers.Adam(hp.Float('Adam_Optim_LR',min_value=1e-4,max_value=1e-2,sampling='LOG',default=1e-3)),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    print("MODEL SUCCESSFULLY CREATED")

    return model
```

```python
hyper_model = MNIST_Fashion_Hyper(imgShape, numClasses)
```

## Set parameters of search space
i.e. Number of trials, epochs per trial, etc.

```python
maxTrials = 20
execPerTrial = 2
numEpochSearch = 10
seedValue = 2

model_tuner = RandomSearch(
    hyper_model,
    objective='val_accuracy',
    seed=seedValue,
    max_trials=maxTrials,
    executions_per_trial=execPerTrial,
    project_name='MNIST_HyperTuned'
)
```

## Summary of tuner
i.e. The hyperparameters that will be tuned



```python
model_tuner.search_space_summary()
```

## Search for best hyperparameters
Note: This code block takes a while to run.

```python
model_tuner.search(x=trainX, y=trainY, epochs=numEpochSearch, validation_data=(valX, valY))
```

```python
model_tuner.results_summary()
```

## Choose best model

```python
best_model = model_tuner.get_best_models()[0]
keras.utils.plot_model(best_model, to_file='HyperTuned_Model_Architecture.png', show_shapes=True, show_layer_names=True)
```

```python
testLoss, testAccuracy = best_model.evaluate(testX, testY)

print("Testing Loss before refitting", testLoss)
print("Testing Accuracy before refitting", "{:.2%}" .format(testAccuracy))
```

## Find optimal number of epochs.

```python
numEpochsToTrain = 20

best_model_history = best_model.fit(trainX, trainY, epochs=numEpochsToTrain, validation_data=(valX, valY),  batch_size=256)

val_acc_per_epoch = best_model_history.history['val_accuracy']
best_epoch = val_acc_per_epoch.index(max(val_acc_per_epoch)) + 1
print('Optimal Number Epochs: %d' % (best_epoch))
```

# Part 5 - Analysis

## Make some predictions to visualize whats happening

```python
n = 5
fig, axes = plt.subplots(1, n, figsize=(25,8))

for i in range(n):
    #Get random image
    randNum = random.randint(0,9999)
    random_image = testX[randNum]
    random_image_class = np.argmax(testY[randNum])
    
    #Make prediction
    prediction = best_hypermodel.predict(random_image.reshape(-1,28,28,1))
    prediction_probability = np.amax(prediction)
    prediction_idx = np.argmax(prediction)

    title = "Actual class = {}\nPredicted class = {}\nPrediction probability = {:.2%}" .format(class_labels[random_image_class], class_labels[prediction_idx], prediction_probability)

    # Print image
    axes[i].imshow(random_image.reshape(28,28), cmap='gray')
    if(prediction_idx == random_image_class):
      axes[i].set_title(title, color='g')
    else:
      axes[i].set_title(title, color='r')
    axes[i].axis('off')

plt.savefig('Prediction_Visualizations.png', dpi=400)    
plt.show()
```

## Confusion Matrix

### Compute prediction accuracy

```python
y_pred, y_true = [], []
for i in range(9000):
  image = testX[i]
  image_class = np.argmax(testY[i])

  prediction = best_hypermodel.predict(image.reshape(-1,28,28,1))
  prediction_idx = np.argmax(prediction)

  y_pred.append(prediction_idx)
  y_true.append(image_class)
```

### Show heatmapped confusion matrix

```python
fig, ax = plt.subplots(figsize=(15,15))
conf_matrix = confusion_matrix(y_true, y_pred, labels=np.arange(10))
conf_matrix = conf_matrix/np.sum(conf_matrix, axis=1)
sns.heatmap(conf_matrix, annot=True, fmt=".2%", square=True, cbar=False, 
            cmap=plt.cm.cividis, xticklabels=classes, yticklabels=classes,
            ax=ax)
ax.set_ylabel('Actual')
ax.set_xlabel('Predicted')
ax.set_title('Confusion Matrix')

plt.savefig('Confusion_Matrix.png', dpi=400)    

plt.show()
```
